```python
import os, json
import pandas as pd
```

```python
path_to_json = '/home/vincenzo/research/eni/paper/twittergeneralinterface/friends_2nd_level_raw_output'

json_files = [pos_json for pos_json in os.listdir(path_to_json) if pos_json.startswith('friends_2nd_level_')]
print("Found",len(json_files),"files")
```

```python
# here I define my pandas Dataframe with the columns I want to get from the json
friends_df = pd.DataFrame(columns=['screen_name', 'location', 'friends_count', 'listed_count', 'followers_count', 'created_at', 'lang', 'statuses_count', 'description'])
links_df = pd.DataFrame(columns=['user_screen_name','screen_name'])
```

```python
# we need both the json and an index number so use enumerate()
cnt_f=0
cnt_l=0
cnt_f_last=0
cnt_l_last=0

screen_names = {}
links = set()
for index, js in enumerate(json_files):
    with open(os.path.join(path_to_json, js)) as json_file:
        print(js, end=" ")
#         cnt_line=0
        
        for line in json_file:
            json_text = json.loads(line)

            # here you need to know the layout of your json and each json has to have
            # the same structure (obviously not the structure I have here)
            user_screen_name = json_text['user_screen_name']
            screen_name = json_text['screen_name']
            
            # insert the friendship relation 
            links.add((user_screen_name, screen_name))
            cnt_l += 1
                        
            # insert only unique friend data 
            if screen_name not in screen_names:         
            
                screen_names[screen_name] = {
                    'location' : json_text['location'],
                    'friends_count' : json_text['friends_count'],
                    'listed_count' : json_text['listed_count'],
                    'followers_count' : json_text['followers_count'],
                    'created_at' : json_text['created_at'],
                    'lang' : json_text['lang'],
                    'statuses_count' : json_text['statuses_count'],
                    'description' : json_text['description']
                }
                cnt_f += 1
            
#             cnt_line+=1
        print("- Loaded\n\tActual amount of links:",cnt_l," (news: ",cnt_l-cnt_l_last,") - Actual unique friends: ",cnt_f)," (news: ",cnt_f-cnt_f_last,")"
        cnt_f_last = cnt_f
        cnt_l_last = cnt_l
```

```python
# filtro i Friend per lingua:

from langdetect import detect_langs

def add_detected_language_column(row):
    """
    Return detected language only if related probability is greater than 90%
    """
    try:
        res = detect_langs(row['description'])
        if res[0].prob>0.9:
            return res[0].lang
    except:
        return None
```
