```python
import pandas
df = pandas.read_csv('EnergyFirstLevel.csv')
df
```

```python
df.groupby("user_screen_name").agg('count')['friends_count']
```

```python
# GLI UTENTI UNIVOCI SONO SOLO 1398 SU 2039
sel_df = df[['screen_name']].drop_duplicates().shape
sel_df
```

```python
# DROP_DUPLICATES LASCIA 1556 UTENTI PERCHE' ALCUNI SONO PRESENTI PIU' VOLTE CON 
# DIFFERENTI VALORI DI followers_count, statuses_count, listed_count E friends_count
friends_df = df.drop(['user_screen_name'], axis=1).drop_duplicates()
friends_df
```

```python
friends_sorted_df = friends_df.sort_values(by=['screen_name','followers_count', 'statuses_count'])
friends_sorted_df
```

```python
# STUDIO LA MOLTEPLICITA' ATTUALE DEGLI UTENTI
multiple_friends_df = friends_sorted_df.groupby("screen_name")\
    .agg('count')[['friends_count']]\
    .sort_values(by=['friends_count'], ascending=False)
multiple_friends_df[multiple_friends_df['friends_count']>1]
```

```python
# PER OGNI UTENTE TENGO SOLO LA RIGA CON PIU' ALTO VALORE DI followers_count
idx = friends_sorted_df.groupby(['screen_name'])['followers_count'].transform(max) == friends_sorted_df['followers_count']
friends_sorted_removeduplbyFC_df = friends_sorted_df[idx]
#friends_sorted_removeduplbyFC_df
```

```python
# PER OGNI UTENTE TENGO SOLO LA RIGA CON PIU' ALTO VALORE DI statuses_count
idx2 = friends_sorted_removeduplbyFC_df.groupby(['screen_name'])['statuses_count'].transform(max) == friends_sorted_removeduplbyFC_df['statuses_count']
friends_sorted_removeduplbyFCandSC_df = friends_sorted_removeduplbyFC_df[idx2]
#friends_sorted_removeduplbyFCandSC_df
```

```python
# STUDIO LA MOLTEPLICITA' ATTUALE DEGLI UTENTI
multiple_friends_sorted_removeduplbyFCandSC_df = friends_sorted_removeduplbyFCandSC_df.groupby("screen_name")\
    .agg('count')[['friends_count']]\
    .sort_values(by=['friends_count'], ascending=False)
multiple_friends_sorted_removeduplbyFCandSC_df[multiple_friends_sorted_removeduplbyFCandSC_df['friends_count']>1]
```

```python
friends_sorted_removeduplbyFCandSC_df[friends_sorted_removeduplbyFCandSC_df['screen_name'] == 'EU_ENV']
```

```python
# PER OGNI UTENTE TENGO SOLO LA RIGA CON PIU' ALTO VALORE DI listed_count
idx3 = friends_sorted_removeduplbyFCandSC_df.groupby(['screen_name'])['listed_count'].transform(max) == friends_sorted_removeduplbyFCandSC_df['listed_count']
friends_sorted_removeduplbyFCandSCandLC_df = friends_sorted_removeduplbyFCandSC_df[idx3]
#friends_sorted_removeduplbyFCandSCandLC_df
```

```python
# STUDIO LA MOLTEPLICITA' ATTUALE DEGLI UTENTI
multiple_friends_sorted_removeduplbyFCandSCandLC_df = friends_sorted_removeduplbyFCandSCandLC_df.groupby("screen_name")\
    .agg('count')[['friends_count']]\
    .sort_values(by=['friends_count'], ascending=False)
multiple_friends_sorted_removeduplbyFCandSCandLC_df[multiple_friends_sorted_removeduplbyFCandSCandLC_df['friends_count']>1]
```

```python
friends_sorted_removeduplbyFCandSCandLC_df[friends_sorted_removeduplbyFCandSCandLC_df['screen_name']=='ItaSIF']
```

```python
# PER OGNI UTENTE TENGO SOLO LA RIGA CON PIU' ALTO VALORE DI friends_count
idx4 = friends_sorted_removeduplbyFCandSCandLC_df.groupby(['screen_name'])['friends_count'].transform(max) == friends_sorted_removeduplbyFCandSCandLC_df['friends_count']
friends_sorted_removeduplbyFCandSCandLCandFRC_df = friends_sorted_removeduplbyFCandSCandLC_df[idx4]
#friends_sorted_removeduplbyFCandSCandLCandFRC_df
```
