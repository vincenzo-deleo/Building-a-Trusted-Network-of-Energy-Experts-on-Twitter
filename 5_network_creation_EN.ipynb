```python
import os, json
import pandas as pd

path_to_json = '/home/vincenzo/research/eni/paper/twittergeneralinterface_bkp-2020.01.27/'
path_to_json2 = '/home/vincenzo/research/eni/paper/twittergeneralinterface/friends_2nd_level_raw_output/'

json_files = [pos_json for pos_json in os.listdir(path_to_json) if pos_json.startswith('friends.')]
json_files2 = [pos_json for pos_json in os.listdir(path_to_json2) if pos_json.startswith('friends_2nd')]
```

```python
len(json_files)
```

```python
len(json_files2)
```

```python
seed_list = pd.read_csv("screen_names_seed.txt")['screen_name'].to_list()
screen_names_1st_level_list = pd.read_csv("screen_names_1st_level_POS_EN.txt")['screen_name'].to_list()
screen_names_2nd_level_list = pd.read_csv("screen_names_2nd_level_EN.txt")['index'].to_list()
```

```python
len(screen_names_1st_level_list)
```

```python
len(screen_names_2nd_level_list)
```

```python
links_1st_level = set()

for index, js in enumerate(json_files):
    with open(os.path.join(path_to_json, js)) as json_file:
        for line in json_file:
            json_text = json.loads(line)
            if json_text['screen_name'] in screen_names_1st_level_list:
                links_1st_level.add((json_text['user_screen_name'],json_text['screen_name']))
```

```python
len(links_1st_level)
```

```python
# COMMENTATO PER EVITARE DI SOVRASCRIVERE ACCIDENTALMENTE I PRECEDENTI RISULTATI

file = open('links_1st_level_EN.csv','w')
for l in links_1st_level:
    file.write(l[0]+","+l[1]+"\n")
file.close()
```

```python
# IL RUN DI QUESTA CELLA RICHIEDE ALCUNE ORE

links_2nd_level = set()

for index, js in enumerate(json_files2):
    with open(os.path.join(path_to_json2, js)) as json_file:
        print(js)
        for line in json_file:
            json_text = json.loads(line)
            if json_text['screen_name'] in screen_names_2nd_level_list:
                links_2nd_level.add((json_text['user_screen_name'],json_text['screen_name']))
```

```python
len(links_2nd_level)
```

```python
# COMMENTATO PER EVITARE DI SOVRASCRIVERE ACCIDENTALMENTE I PRECEDENTI RISULTATI

file = open('links_2nd_level_EN.csv','w')
for l in links_2nd_level:
    file.write(l[0]+","+l[1]+"\n")
file.close()
```
